```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
merged_data= pd.read_csv(r"D:\plot bacteria\sanatlucia_merged_data.csv") 
final_df= pd.read_csv(r"D:\plot bacteria\final_df_santalucia.csv")
ganji_merged_data=pd.read_csv(r"D:\plot bacteria\ganji_merged_data 1.csv") 
ganji_final_df=pd.read_csv(r"D:\plot bacteria\ganji_final_df.csv")
sponer_merged_data=pd.read_csv(r"D:\plot bacteria\sponer_merged_data 1.csv") 
sponer_final_df=pd.read_csv(r"D:\plot bacteria\sponer_final_df.csv") 
ussrey_merged_data=pd.read_csv(r"D:\plot bacteria\ussrey_merged_data.csv") 
ussrey_final_df=pd.read_csv(r"D:\plot bacteria\ussrey_final_df.csv") 
perez_merged_data=pd.read_csv(r"D:\plot bacteria\perez_merged_data.csv") 
perez_final_df=pd.read_csv(r"D:\plot bacteria\perez_final_df.csv") 
perez_RNA_merged_data=pd.read_csv(r"D:\plot bacteria\perez(RNA)_merged_data.csv") 
perez_RNA_final_df=pd.read_csv(r"D:\plot bacteria\perez(RNA)_final_df.csv") 
anselmi_merged_data=pd.read_csv(r"D:\plot bacteria\anselmi_merged_data.csv") 
anselmi_final_df=pd.read_csv(r"D:\plot bacteria\anselmi_final_df.csv") 
```

```python
# setting the figure size and width ratios
fig, axs = plt.subplots(7, 12, figsize=(60, 28), gridspec_kw={'width_ratios': [6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6]})

# define a list of tuples containing merged dataframes and their corresponding final dataframes
dataframes = [
    (merged_data, final_df),
    (ganji_merged_data, ganji_final_df),
    (sponer_merged_data, sponer_final_df),
    (ussrey_merged_data, ussrey_final_df),
    (perez_merged_data, perez_final_df),
    (perez_RNA_merged_data, perez_RNA_final_df),
    (anselmi_merged_data, anselmi_final_df),
]

# iterate through the dataframes list and create subplots for each one
for i, df in enumerate(dataframes):
    # set the common x-axis for this dataframe
    x = df[0]['rolling_avg_position']
    for j, col in enumerate(df[0].columns[:-1]):
        # plotting the data
        axs[i][j].plot(x, df[0][col], color='darkslategrey')
        axs[i][j].plot(df[1]['Index'], df[1][col], color='orange')
        # setting the title and axes labels
        axs[i][j].set_title(col)
        axs[i][j].set_xlabel('Nucleotide Position')
        axs[i][j].set_ylabel('Values from Santalucia et. al')

# adjusting the spacing between subplots
plt.tight_layout()

# displaying the plot
plt.show()
```

```python
# setting the figure size and width ratios
fig, axs = plt.subplots(7, 12, figsize=(60, 28), gridspec_kw={'width_ratios': [6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6]})

# define a list of tuples containing merged dataframes and their corresponding final dataframes
dataframes = [
    (merged_data, final_df),
    (ganji_merged_data, ganji_final_df),
    (sponer_merged_data, sponer_final_df),
    (ussrey_merged_data, ussrey_final_df),
    (perez_merged_data, perez_final_df),
    (perez_RNA_merged_data, perez_RNA_final_df),
    (anselmi_merged_data, anselmi_final_df),
]

# define a list of y-axis labels for each row
ylabels = ['Santalucia et. al', 'Ganji et al', 'Sponer et al', 'Ussrey et al', 
           'Perez et al', 'Perez(RNA) et al', 'Anselmi et al']

# iterate through the dataframes list and create subplots for each one
for i, df in enumerate(dataframes):
    # set the common x-axis for this dataframe
    x = df[0]['rolling_avg_position']
    for j, col in enumerate(df[0].columns[:-1]):
        # plotting the data
        axs[i][j].plot(x, df[0][col], color='crimson',linewidth=3)
        axs[i][j].plot(df[1]['Index'], df[1][col], color='gold')
        # setting the title and axes labels
        axs[i][j].set_title(col, fontsize=20)
        axs[i][j].set_xlabel('Nucleotide Position', fontsize=16)
        axs[i][j].set_ylabel(ylabels[i], fontsize=20)

# adjusting the spacing between subplots
plt.tight_layout()

# displaying the plot
plt.show()
```
